### Geographical Coordinates

The geographical coordinates for London data has been obtained from the GeoPy library in python. This data is relevant for plotting the map of London using the Folium library in python. The geocoder library in python has been used to obtain latitude and longitude data for various neighborhoods in London. The coordinates of all neighborhoods in London are used to check the accuracy of coordinates given on Wikipedia and replace them in our data frame if the absolute difference is more than 0.001. These coordinates are then further used for plotting using the Folium library in python.
